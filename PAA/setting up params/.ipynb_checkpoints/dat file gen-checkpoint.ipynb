```python
O = {'Atom':'O1-','Charge':-0.635,'Epsilon':6.900176589,'Radius':0.29474215}
CH = {'Atom':'C','Charge':0,'Epsilon':0.379557318,'Radius':0.563384105}
CH2 = {'Atom':'C','Charge':0,'Epsilon':1.642169424,'Radius':0.456884852}
```

```python
charges = []
atoms = []
epsilons = []
residues = []
radii = []
names = []
for row in df.iterrows():
    atoms.append(row[1][12])
    if row[1][11] == 'C':
        names.append('C')
        if row[1][12]%3 == 2:
            
            charges.append(CH['Charge'])
            epsilons.append(CH['Epsilon'])
            radii.append(CH['Radius'])
            residues.append(row[1][2])
        elif row[1][12]%3 == 1:
            #atoms.append(CH2['Atom'])
            charges.append(CH2['Charge'])
            epsilons.append(CH2['Epsilon'])
            radii.append(CH2['Radius'])
            residues.append(row[1][2])
        else:
            #atoms.append(carboxyl['Atom'])
            charges.append(carboxyl['Charge'])
            epsilons.append(carboxyl['Epsilon'])
            radii.append(carboxyl['Radius'])
            residues.append(row[1][2])
    else:
        names.append('O')
        #atoms.append(O['Atom'])
        charges.append(O['Charge'])
        epsilons.append(O['Epsilon'])
        radii.append(O['Radius'])
        residues.append(row[1][2])
```

```python
X = pd.DataFrame(data=np.transpose(np.asarray([residues,atoms,charges,radii,epsilons,names])),columns=X.columns)
```

<h3>Insert a Sodium ion for each pair of O1- atoms

```python
ions = pd.DataFrame(columns=df.columns)
j = 0
for i in range(len(df)):
    if df.iloc[i,11] == 'O1-':
        if ((i-1)>=0) & (df.iloc[i-1,11] == 'O1-'):
            row = df.loc[i].copy()
            row['Type'] = 'HETATM'
            row['x'] = row['x']*0.975
            row['y'] = row['y']*0.975
            row['z'] = row['z']*0.975
            row['atom_name'] = 'NA'
            row['Residue_name'] = 'NA'
            row['atom_symbol'] = 'NA+'
            row['atom number'] = 151+j
            ions.loc[j] = row
            j += 1
```
